```python
#29. Create a new column popularity_score as views_millions + likes_millions.

df["popularity_score"] = df["views_millions"] + df["likes_millions"]
df.head()
```

```python
#30. Create a new column rating_category using apply().

def category_rating(x):
    if x >= 8:
        return "Excellent"
    elif x >= 6:
        return "Good"
    elif x >= 4:
        return "Average"
    else:
        return "Poor"

df['rating_category'] = df['imdb_rating'].apply(category_rating)
df.head()
```

```python
#31. Map subscription_plan to short codes.

plan_code = {"Basic" : "B", "Standard" : "S", "Mobile" : "M", "Premium" : "P"}

df["subscription_short_codes"] = df["subscription_plan"].map(plan_code)
df.head()
```

```python
#32. Map is_trending values to numerical format.

trending_value = {"No" : 0, "Yes" : 1}

df["trending_value"] = df["is_trending"].map(trending_value)
df.head()
```
